```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/blob/main/uybor_scrapping.csv?raw=true")
df.head()
```

# Uylarning narxi taqsimotini grafikda ko'rsating

```python

df = df[df['price']<200000]
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='price')
plt.savefig('graph6.png')
plt.show()
```

# Uylarning maydoni taqsimoti grafigini chizing

```python

df = df[df['price']<200000]
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='size')
plt.savefig('graph7.png')
plt.show()
```

# Uylarning xonalar soni taqsimotini chiqaring.

```python

df = df[df['price']<200000]
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='rooms')
plt.savefig('graph8.png')
plt.show()
```

# Uylarning narxi taqsimotini xonalar soniga bog'lab chiqaramiz.

```python
df = df[df['rooms'] < 7]
sns.displot(data=df, x='price', hue='rooms', palette='pastel', height=5, aspect=2)
plt.savefig('graph9.png')
plt.show()
```

# Faqatgina 1,2,3 xonali uylarning narx bo'yicha taqsimotini chizamiz.

```python
df123 = df[df.rooms <= 3]
sns.displot(data=df123, x='price', hue='rooms', palette='dark', height=5, aspect=2)    
plt.savefig('graph10.png')
plt.show()
```

# 4,5,6 xonali uylar uchun narx taqsimotini alohida grafiklarda chizing

```python
df456 = df[df.rooms.between(4,6)]
sns.displot(data=df456, x='price', col='rooms', height=5, aspect=1)
plt.savefig('graph11.png')
plt.show()
```

# Uylarning maydoni taqsimoti KDE grafigini chizing

```python
plt.figure(figsize=(12,20))
sns.displot(data=df, kind='kde', x='size')
plt.tight_layout()
plt.savefig('graph12.png')
plt.show()
```

# Uylarning maydoni taqsimotini KDE va taqsimotini birlashtirib chizing.

```python
plt.figure(figsize=(12, 10))

sns.displot(data=df, kde=True, x='size', height=5, aspect=2)
plt.savefig('graph13.png')
plt.show()
```

# Uylarning narxi va xonalari soni uchun (2,3,5 xonali) KDE chizing.

```python
plt.figure(figsize=(12, 6))

sns.kdeplot(data=df[df.rooms.isin([2, 3, 5])],  x='price', hue='rooms',  shade=True)
plt.savefig('graph14.png')
plt.show()
```

# Uylarning maydoni uchun kumulyativ taqsimot grafigini chizing (CDF)


```python
plt.figure(figsize=(12, 6))
sns.ecdfplot(data=df, x='size')
plt.grid(which='both')
plt.minorticks_on()
plt.savefig('graph15.png')
plt.show()
```

# Yuqoridagi grafikni uylarning narxi uchun chizing.

```python
plt.figure(figsize=(12, 6))
sns.ecdfplot(data=df, x='price')
plt.grid(which='both')
plt.minorticks_on()
plt.xticks(rotation=90)
plt.savefig('graph16.png')
plt.show()
```
