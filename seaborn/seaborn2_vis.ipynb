```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\EcoRestart\Desktop\ds-practicum\matplotlib\Updated_climate.csv", index_col=0)
df
```

```python
legends = ["Yillik o'rtacha harorat", "Yillik o'rtacha maksimal harorat", "Yillik o'rtacha minimum harorat", "Yillik o'rtacha shamol tezligi", "Yomg'irli kunlar soni", "Bo'ronli kunlar soni", "Tumanli kunlar soni", "To'fonli kunlar soni", "Do'l yoqqan kunlar"]
plt.figure(figsize=(16, 10))

sns.lineplot(data=df)
plt.legend(legends)
plt.title("Toshkent shaxri uchun ob-havo ko'rsatkichlari")
plt.xlabel('Yil')
plt.ylabel('Harorat')
plt.grid(which='both')
plt.savefig('graph3.png')
plt.show()
```

# Haroratni ko'rsatuvchi ustunlarni alohida chiziqli grafik qilib chiqaring.

```python
dfT = df[['T', 'TM', 'Tm']]
plt.figure(figsize=(15, 9))
legends = ["Yillik o'rtacha harorat", "Yillik o'rtacha maksimal harorat", "Yillik o'rtacha minimum harorat"]
sns.lineplot(dfT)
plt.legend(legends)
plt.title("Toshkent shaxri uchun ob-havo ko'rsatkichlari")
plt.xlabel('Yil')
plt.ylabel('Harorat')
plt.grid()
plt.tight_layout()
plt.savefig('graph2.png')
plt.show()
```

# Kunlarni ko'rsatuvchi ustunlarni alohida grafik ko'rinishida chiqaring

```python
plt.figure(figsize=(15, 9))

legends = ["Yomg'irli kunlar soni", "Bo'ronli kunlar soni", "Tumanli kunlar soni", "To'fonli kunlar soni", "Do'l yoqqan kunlar"]
dfDay = df[['SN', 'TS', 'FG', 'TN', 'GR']]

sns.lineplot(dfDay)
plt.legend(legends)
plt.title("Toshkent shaxri uchun ob-havo ko'rsatkichlari")
plt.xlabel('Yil')
plt.ylabel('Harorat')
plt.grid()
plt.minorticks_on()
plt.tight_layout()
plt.savefig('graph1.png')
plt.show()
```

# Yillik o'rtacha shamol tezligini ustunli grafik ko'rinishida chiqaring.

```python
dfwind = df['V']

plt.figure(figsize=(15, 5))

sns.barplot(data=dfwind, palette='bright')
plt.xticks(rotation=90)
plt.title("Toshkent shaxri uchun ob-havo ko'rsatkichlari")
plt.legend(["Yillik o'rtacha shamol tezligi"])
plt.xlabel("Yil")
plt.ylabel("Harorat")
plt.savefig('graph4.png')
plt.show()
```

# Yomg'rili, qorli, bo'ronli, tumanli va do'l yoqqan kunlar sonini heatmap ko'rinishida chiqaring.

```python
legends = ["Yomg'irli kunlar soni", "Bo'ronli kunlar soni", "Tumanli kunlar soni", "To'fonli kunlar soni", "Do'l yoqqan kunlar"]
dfDay = df[['SN', 'TS', 'FG', 'TN', 'GR']]
yticks_labels = ["Yomg'ir", "Bo'ron", "Tuman", "To'fon", "Do'l"]

plt.figure(figsize=(30, 8))

sns.heatmap(data=dfDay.T, cmap='YlGnBu', linewidths=2,  square=True)
plt.title("Toshkent shaxri uchun ob-havo ko'rsatkichlari")
plt.xlabel('Yil')
plt.yticks(np.arange(5) + 0.5, labels=yticks_labels)
plt.yticks(rotation=0)
plt.xticks(rotation=90)
plt.savefig('graph5.png')
plt.show()
```
