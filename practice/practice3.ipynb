# Barcha ma'lumotlar to'plamini kuzatgan bo'lsangiz, ba'zilarining ustunlari o'zgarib kelmoqda. Lekin bu bizga qiziq emas. Biz asosiy maqsadni ushbu ma'lumotlardan faqat mamlakatlarning baxtlilik darajalarinigina ajratib olishga qaratamiz.
# Vazifa: Barcha ma'lumotlar to'plamlaridan faqat mamalakatlar (Country yoki Country or region), ularning baxtlilik darajasini (Happiness Rank, Happiness.Rank yoki Overall rank) va baholarini (Happiness Score, Happiness.Score yoki Score) ajratib oling.

```python
def extractor(dataframe):
    columns = ['Country', 'Country or region', 'Happiness Rank', 'Happiness.Rank', 'Overall rank', 'Happiness Score', 'Happiness.Score', 'Score']
    mask = [col for col in dataframe.columns if col in columns]
    return dataframe[mask]

dataframes = [data15, data16, data17, data18, data19]
data15, data16, data17, data18, data19 = [extractor(dataframe) for dataframe in dataframes]
```

# Navbatdagi mashqimiz irarxik ustunlar haqida bo'ladi.
# Vazifa: Barcha ma'lumotlar to'plamini ustunlarini irarxik ko'rinishga keltirish kerak. Quyidagi rasmda 2015 yil uchun qanday ko'rinishda bo'lishi ko'rsatilgan. Qolganlarini mustaqil bajaring.

```python
dataframes = [ data15, data16, data17, data18, data19]
years = [ 2015, 2016, 2017, 2018, 2019]

for i in range(len(dataframes)):
    df = dataframes[i]
    year = years[i]
    df.columns = pd.MultiIndex.from_product([[year], df.columns])
    dataframes[i] = df
```

# Barcha ma'lumotlarni irarxik ustunlar ko'rinishiga keltirib olgan bo'lsangiz, unda navbatdagi mashqimizda ularni birlashtiramiz.
# Vazifa: Gorizantal o'q ko'rinishida barcha ma'lumotlar to'plamlarini birlashtiring (axis=1). Va natijaviy dataframe ustunlariga Years va Info deb nom bering.
